# Creating Plots

### Fuctional Method

```python
# Create Plots
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() # don't need for Jupyter notebooks
```

```python
# Create multi plots on the same canvas
plt.subplot(1,2,1) # num rows, columns, plot number
plt.plot(x, y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object Oriented Method

```python
# Create empty figure object
fig = plt.figure()

# Add axes
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height of axis from 0 to 1
axes2 = fig.add_axes([0.2,.5,.4,.30])

# Plot on axis
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

# **Creating Subplots**
Two methods to create subplots:  
plt.figure()  
plt.subplot()  

Key difference:   
plt.subplot adds axes automatically based on the number of rows and columns specified

### Method 1: plt.figure()

```python
# Create empty figures and axes
fig, axes = plt.subplots(nrows = 1, ncols =2) #tuple unpacking

# Fix overlaps
plt.tight_layout()

# Iterate through through the list of mabplotlib axes objects to plot
for current_ax in axes: 
    current_ax.plot(x,y)
```

### Method 2: plt.subplots()

```python
# Create empty figures and axes
fig, axes = plt.subplots(nrows = 1, ncols =2) #tuple unpacking

# Index to plot on the first axis
axes[0].plot(x,y)
axes[0].set_title('First Plot')

# Index to plot on the second axis
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

# Fix overlaps
plt.tight_layout()
```

# Figure Size and DPI (Dots Per Inch)

```python
# Set figure size using plt.figure
fig = plt.figure(figsize = (8,2)) # tuple specifying height and width inches

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Set fig size using plt.subplots
fig,axes = plt.subplots(nrows = 2, ncols = 1,figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

# Legends

```python
# Create Figure
fig = plt.figure()

# Create axes
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared')
ax.plot(x,x**3, label ='X Cubed')

# Add legend at best location
ax.legend(loc = 0)

# Add legend using tuple
ax.legend(loc = (0.1,0.1))
```

# Plot Appearance

```python
# Create plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Customize line 
ax.plot(x,y, color = '#FF8C00', lw = 3, alpha = 0.5, ls = '--', marker ='o', markersize=20,
       markerfacecolor = 'green',markeredgewidth = 3, markeredgecolor = 'blue') # RGB Hex Code
```

# Axis Appearance

```python
# Create plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Customize line 
ax.plot(x,y, color = 'purple', lw = 2,ls = '--', marker ='o', markersize=10)

# Set upper and lower bound for x and y axis
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
